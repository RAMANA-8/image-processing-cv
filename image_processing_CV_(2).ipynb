```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('img.jpeg')
plt.imshow(img)
```

```python
cv2.imwrite('img.jpeg',cv2.Canny(img,200,300))
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'gray')
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
img_c = img.copy()
# Draw a rectangle
cv2.rectangle(img_c, pt1 = (500,100), pt2 = (200, 80),
              color = (1000, 1000, 0), thickness =100)
plt.imshow(img_c)
```

```python
cv2.circle(img_c, center = (50, 50), radius = 50,
           color = (0, 0, 255), thickness = 5)
plt.imshow(img_c)
```

```python
cv2.putText(img_c, text = "It is easy ",
            org = (0, 150),
            fontFace = cv2.FONT_HERSHEY_DUPLEX,
            fontScale = 2,
            color = (0, 255, 0),
            thickness = 2,
            lineType = cv2.LINE_AA)
plt.imshow(img_c)
```

```python
from PIL import Image

img = Image.open("img.jpeg")

img.show()
```

```python
from PIL import Image, ImageFilter

# Open an image file
with Image.open('img.jpeg') as img:
    # Apply a filter (e.g., blur)
    blurred_img = img.filter(ImageFilter.BLUR)

    # Save the processed image
    blurred_img.save('output.jpg')

    # Optionally, show the image
    blurred_img.show()
```

```python
# Resize image
resized_img = img.resize((800, 600))
resized_img.save('resized.jpg')
resized_img.show()
```

```python
# Crop image
cropped_img = img.crop((100, 100, 400, 400))
cropped_img.save('cropped.jpg')
cropped_img.show()
```

```python
# Convert to grayscale
grayscale_img = img.convert('L')
grayscale_img.save('grayscale.jpg')
grayscale_img.show()
```
